```python
# Activity 1 - From Raw Text to Tokens

# Activity 1.1
sentence = "The Battery life on this Phone is AMAZING... it lasts 2 full days! :)"

# TODO 1
lower_sentence = sentence.lower()
print("TODO 1 - lower_sentence:")
print(lower_sentence)
print()

# TODO 2
clean_chars = []
for char in lower_sentence:
    if char.isalpha() or char.isdigit() or char == " ":
        clean_chars.append(char)
clean_sentence = "".join(clean_chars)

print("TODO 2 - clean_sentence:")
print(clean_sentence)
print()

# TODO 3
def clean_text(text):
    result = []
    for char in text.lower():
        if char.isalpha() or char.isdigit() or char == " ":
            result.append(char)
    return "".join(result)

print("TODO 3 - clean_text tests:")
test_sentences = [
    "I LOVE this! Best product ever.",
    "Terrible... absolutely terrible. 0/10.",
    "Not bad :) would buy again!",
]
for s in test_sentences:
    print(f"IN : {s}")
    print(f"OUT: {clean_text(s)}")
    print()

# Activity 1.2
raw_sentence = "I love NLP... and Python! It's amazing :)"

# TODO 4
raw_tokens = raw_sentence.split(" ")
print("TODO 4 - raw_tokens:")
print(raw_tokens)
print()

# TODO 5
clean_tokens = clean_text(raw_sentence).split()
print("TODO 5 - clean_tokens:")
print(clean_tokens)
print()

# TODO 6
tokens_single_space = clean_text(raw_sentence).split(" ")
tokens_any_space = clean_text(raw_sentence).split()
print('TODO 6 - split(" "):', tokens_single_space)
print('TODO 6 - split()   :', tokens_any_space)
print()

# Activity 1.3
# TODO 7
def tokenise_basic(text):
    tokens = []
    buffer = ""

    for char in text.lower():
        if char.isalpha() or char.isdigit():
            buffer += char
        elif char == " ":
            if buffer:
                tokens.append(buffer)
                buffer = ""
        else:
            pass

    if buffer:
        tokens.append(buffer)

    return tokens

# TODO 8
EMOTICONS = {":)", ":(", ":D", ":P", ";)", ":/", ":o", "XD", "<3"}

def tokenise_with_emoticons(text):
    tokens = []
    buffer = ""
    i = 0
    text = text.lower()
    emoticons_lower = {e.lower() for e in EMOTICONS}

    while i < len(text):
        two_chars = text[i:i+2]
        if two_chars in emoticons_lower:
            if buffer:
                tokens.append(buffer)
                buffer = ""
            tokens.append(two_chars)
            i += 2
            continue

        char = text[i]
        if char.isalpha() or char.isdigit():
            buffer += char
        elif char == " ":
            if buffer:
                tokens.append(buffer)
                buffer = ""
        else:
            pass

        i += 1

    if buffer:
        tokens.append(buffer)

    return tokens

print("TODO 7 - tokenise_basic:")
print(tokenise_basic(raw_sentence))
print()
print("TODO 8 - tokenise_with_emoticons:")
print(tokenise_with_emoticons(raw_sentence))
print()

# TODO 9
def tokenise(text, keep_emoticons=True):
    if keep_emoticons:
        return tokenise_with_emoticons(text)
    return tokenise_basic(text)

print("TODO 9 - configurable tokeniser:")
s_bonus = "Great experience :) would totally recommend!"
print(tokenise(s_bonus, keep_emoticons=True))
print(tokenise(s_bonus, keep_emoticons=False))
print()

# Activity 1.4
EMOTICONS_14 = {
    ":)", ":(", ":D", ":P", ";)", ":/", ":o", "XD", "<3",
    ":-)", ":-(", ":-D", ":-P"
}

STOPWORDS = {
    "a", "an", "the",
    "in", "on", "at", "to", "for", "of", "with", "by", "from",
    "and", "but", "or", "so",
    "is", "are", "was", "were", "be", "been", "being",
    "do", "does", "did",
    "have", "has", "had",
    "will", "would", "could", "should",
    "i", "you", "he", "she", "it", "we", "they",
    "me", "him", "her", "us", "them",
    "my", "your", "his", "its", "our", "their",
    "this", "that", "these", "those",
    "just", "also", "very",
}

def lowercase_14(text):
    return text.lower()

# TODO 10
def tokenise_14(text, keep_emoticons=True):
    tokens = []
    buffer = ""
    i = 0
    text = text.lower()
    emoticons_lower = {e.lower() for e in EMOTICONS_14}

    while i < len(text):
        if keep_emoticons:
            three_chars = text[i:i+3]
            two_chars = text[i:i+2]
            if three_chars in emoticons_lower:
                if buffer:
                    tokens.append(buffer)
                    buffer = ""
                tokens.append(three_chars)
                i += 3
                continue
            if two_chars in emoticons_lower:
                if buffer:
                    tokens.append(buffer)
                    buffer = ""
                tokens.append(two_chars)
                i += 2
                continue

        char = text[i]
        if char.isalpha() or char.isdigit():
            buffer += char
        elif char == " ":
            if buffer:
                tokens.append(buffer)
                buffer = ""
        else:
            pass

        i += 1

    if buffer:
        tokens.append(buffer)
    return tokens

# TODO 11
def remove_stopwords(tokens, stopwords=STOPWORDS):
    return [token for token in tokens if token not in stopwords]

# TODO 12
def remove_short_tokens(tokens, min_length=2):
    emoticons_lower = {e.lower() for e in EMOTICONS_14}
    return [token for token in tokens if len(token) >= min_length or token in emoticons_lower]

# TODO 13
def pipeline(text, keep_emoticons=True, apply_stopwords=True, min_length=2):
    text = lowercase_14(text)
    tokens = tokenise_14(text, keep_emoticons=keep_emoticons)
    if apply_stopwords:
        tokens = remove_stopwords(tokens)
    tokens = remove_short_tokens(tokens, min_length=min_length)
    return tokens

print("Activity 1.4 - pipeline tests:")
test_cases_14 = [
    "The battery life on this phone is absolutely amazing! :)",
    "I can't believe how terrible this product is. Total waste of money.",
    "Not bad, not great... just okay.",
    "Incredible! Best purchase I've made all year :D",
    "Rude staff, cold food. Never coming back.",
]
for sentence in test_cases_14:
    print(sentence)
    print(pipeline(sentence))
    print()

sentence_14 = "This product is NOT as good as I expected. Very disappointing :("
v1 = pipeline(sentence_14, keep_emoticons=True, apply_stopwords=True)
v2 = pipeline(sentence_14, keep_emoticons=True, apply_stopwords=False)
v3 = pipeline(sentence_14, keep_emoticons=False, apply_stopwords=True)
v4 = pipeline(sentence_14, keep_emoticons=False, apply_stopwords=False)

print("Activity 1.4 - variants:")
print("V1:", v1)
print("V2:", v2)
print("V3:", v3)
print("V4:", v4)
```

```python
# Activity 3 and Activity 4 - Vocabulary, BoW, TF, IDF, TF-IDF

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# Session 3 corpus from the notebook
documents = [
    ["film", "amazing", "battery", "lasts", "two", "days", "amazing"],
    ["screen", "bad", "scratches", "easily", "bad", "product"],
    ["amazing", "product", "highly", "recommend", "recommend"],
    ["battery", "lasts", "days", "great", "product"],
    ["screen", "great", "display", "film", "amazing"],
]
labels = ["positive", "negative", "positive", "positive", "positive"]

# Activity 3.1
# TODO 1-3
def build_vocabulary(documents):
    all_tokens = {token for doc in documents for token in doc}
    vocabulary = sorted(all_tokens)
    word_to_idx = {word: i for i, word in enumerate(vocabulary)}
    return vocabulary, word_to_idx

vocabulary, word_to_idx = build_vocabulary(documents)
vocab_list = vocabulary
print("Activity 3.1 - vocabulary:")
print(vocabulary)
print("Vocabulary size:", len(vocabulary))
print("'amazing' ->", word_to_idx.get("amazing"))
print()

# Activity 3.2
# TODO 4-5
def vectorise(tokens, word_to_idx):
    vector = np.zeros(len(word_to_idx), dtype=int)
    for token in tokens:
        if token in word_to_idx:
            vector[word_to_idx[token]] += 1
    return vector

vec0 = vectorise(documents[0], word_to_idx)
print("Activity 3.2 - vec0:")
print(vec0)
print()

vec_a = vectorise(["not", "good"], word_to_idx)
vec_b = vectorise(["good"], word_to_idx)
print("Negation experiment:")
print("'not good':", vec_a)
print("'good'    :", vec_b)
print("Identical?:", np.array_equal(vec_a, vec_b))
print()

# Activity 3.3
# TODO 6
def build_dtm(documents, word_to_idx):
    matrix = np.array([vectorise(doc, word_to_idx) for doc in documents])
    return matrix

dtm = build_dtm(documents, word_to_idx)
print("Activity 3.3 - DTM shape:", dtm.shape)
zeros = np.sum(dtm == 0)
total = dtm.size
print(f"Sparsity: {zeros / total * 100:.1f}% zeros")
print()

# TODO 7
print("Most frequent token per document:")
for i, row in enumerate(dtm):
    top_word = vocabulary[np.argmax(row)]
    print(f"Doc {i} [{labels[i]}]: {top_word}")
print()

# Activity 4.1
# TODO 8
def compute_tf(tokens, word_to_idx):
    counts = np.zeros(len(word_to_idx), dtype=float)
    for token in tokens:
        if token in word_to_idx:
            counts[word_to_idx[token]] += 1
    doc_length = len(tokens)
    tf = counts / doc_length
    return tf

tf_matrix = np.array([compute_tf(doc, word_to_idx) for doc in documents])
print("Activity 4.1 - TF matrix shape:", tf_matrix.shape)
print("Value range:", tf_matrix.min(), "to", tf_matrix.max())
print()

# TODO 9
print("Top-3 TF tokens per document:")
for i, row in enumerate(tf_matrix):
    top3_idx = np.argsort(row)[::-1][:3]
    top3 = [(vocabulary[j], float(row[j])) for j in top3_idx]
    print(f"Doc {i}: {top3}")
print()

# Activity 4.2
# TODO 10-11
def compute_idf(dtm):
    N = dtm.shape[0]
    df = np.sum(dtm > 0, axis=0)
    idf = np.log(N / df)
    return idf

idf = compute_idf(dtm)
sorted_idx = np.argsort(idf)
print("Lowest IDF terms:")
for i in sorted_idx[:5]:
    print(vocab_list[i], round(float(idf[i]), 4))
print()
print("Highest IDF terms:")
for i in sorted_idx[-5:][::-1]:
    print(vocab_list[i], round(float(idf[i]), 4))
print()

# Activity 4.3
# TODO 12
tfidf_matrix = tf_matrix * idf
print("Activity 4.3 - TF-IDF matrix shape:", tfidf_matrix.shape)
print()

# TODO 13
print("Top-3 TF-IDF tokens per document:")
for i, row in enumerate(tfidf_matrix):
    top3_idx = np.argsort(row)[::-1][:3]
    top3 = [(vocabulary[j], round(float(row[j]), 4)) for j in top3_idx]
    print(f"Doc {i} [{labels[i]}]: {top3}")
print()

# sklearn comparison
corpus_str = [' '.join(doc) for doc in documents]
sk_vec = TfidfVectorizer()
X_sk = sk_vec.fit_transform(corpus_str).toarray()
sk_vocab = sk_vec.get_feature_names_out()

print("sklearn comparison for doc 0:")
print("Our basic TF-IDF top-3:")
for j in np.argsort(tfidf_matrix[0])[::-1][:3]:
    print(f"  {vocab_list[j]} -> {tfidf_matrix[0, j]:.4f}")
print("sklearn top-3:")
for j in np.argsort(X_sk[0])[::-1][:3]:
    print(f"  {sk_vocab[j]} -> {X_sk[0, j]:.4f}")
```

## Activity 3 and 4 - Written Answers

### Activity 3.1 Discussion

If a new document contains a word not in the vocabulary, that word is Out of Vocabulary (OOV). In this simple implementation it is ignored, so the model loses that information.

### Activity 3.2 Reflection

The vectors for `"not good"` and `"good"` are identical in this toy example because neither token appears in the corpus vocabulary. More generally, even when both words exist, Bag of Words is still weak at negation because it does not model word order or composition.

### Activity 4.1 Interpretation

TF removes document-length bias by dividing counts by the number of tokens in each document. That makes short and long documents more comparable.

### Activity 4.2 Interpretation

IDF downweights terms that appear in many documents and upweights terms that are rare across the corpus. Common words have low IDF. Distinctive words have high IDF.

### Activity 4.3 sklearn Comparison

The rankings should be similar, but the exact values differ because `TfidfVectorizer` uses:

- smoothed IDF: `log((1 + N) / (1 + df)) + 1`
- L2 normalization by default

So it is normal for the numbers not to match exactly.

### Closing Question

Classical methods like BoW and TF-IDF have a clear ceiling: they treat text mostly as token counts, so they miss word order, context, negation, semantic similarity, and nuance. Two reviews with opposite meanings can look very similar if they share many of the same words.

The field moved to embeddings because embeddings represent words and documents in a continuous semantic space, which captures similarity and context much better than simple count-based methods.
